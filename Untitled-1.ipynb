```python
#teste do metodo chi2_contingency
df_contigencia = pd.crosstab(df['Contract'], df.Churn)
p_valor = chi2_contingency(df_contigencia).pvalue
p_valor
```

```python
#calculo de pvalor em todo cat_col
for teste in cat_col:
    cr = pd.crosstab(df[teste], df.Churn)
    pvalor = chi2_contingency(cr).pvalue
    
    print(f'coluna: {teste} p_valor = {pvalor}')
    if pvalor < 0.05:
        print(f'rejeitamos a hipotese nula, há interferencia estatistica em coluna {teste} e Churn')
    else:
        print(f'aceitamos a hipotese nula, nao há interferencia estatistica em coluna {teste} e Churn')
```

```python
#agora vamos fazer o teste estatistico para as colunas numericas 
for num in num_col:
    cr_num = pd.crosstab(df[num], df.Churn)
    p_valor = chi2_contingency(cr_num).pvalue
    print(f'coluna {num} e pvalor = {p_valor}')
    if p_valor < 0.05:
        print(f'rejeitamos a hipotese nula, há interferencia estatistica em coluna {num} e Churn')
    else:
        print(f'aceitamos a hipotese nula, nao há interferencia estatistica em coluna {num} e Churn')
```

optuna

```python
#teste de hiperparametros para xgboost
def objective_xgboost(trial):
    
    learning_rate = trial.suggest_float("learning_rate", 0.01, 0.3, log=True)
    n_estimators = trial.suggest_int("n_estimators", 50, 300)
    max_depth = trial.suggest_int("max_depth", 3, 10)

    
    modelo_xgboost.set_params(
        learning_rate=learning_rate,
        n_estimators=n_estimators,
        max_depth=max_depth
    )

    
    modelo_xgboost.fit(x_treino, y_treino)

    predictions = modelo_xgboost.predict(x_teste)

    score = accuracy_score(y_teste, predictions)

    return 1.0 - score  # Minimizando o erro
```

```python
#teste de hiperparametros para arvore de decisao
def objective_arvore(trial):
    
    max_depth = trial.suggest_int("max_depth", 1, 30)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 20)
    max_features = trial.suggest_categorical("max_features", [None, "sqrt", "log2"])
    criterion = trial.suggest_categorical("criterion", ["gini", "entropy"])
    splitter = trial.suggest_categorical("splitter", ["best", "random"])
    
    
    model = DecisionTreeClassifier(
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        criterion=criterion,
        splitter=splitter,
        random_state=42
    )
    
    model.fit(x_treino, y_treino)
    predictions = model.predict(x_teste)
    score = accuracy_score(y_teste, predictions)
    return 1.0 - score  # Minimiza o erro
```

```python
study_arvore = optuna.create_study(direction="minimize")
study_arvore.optimize(objective_arvore, n_trials=50)
```

```python
print("Melhores hiperparâmetros arvore:", study_arvore.best_params)
print("Melhor valor (erro):", study_arvore.best_value)
```

```python

study_xgboost = optuna.create_study(direction="minimize")  # Ou "maximize" dependendo do problema
study_xgboost.optimize(objective_xgboost, n_trials=50)  # 50 tentativas de hiperparâmetros
```

```python
print("Melhores hiperparâmetros modelo_xgboost:", study_xgboost.best_params)
print("Melhor valor (erro):", study_arvore.best_value)
```
