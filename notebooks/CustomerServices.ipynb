```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data_path = os.path.join('..', 'src', 'analytics', 'CustomerServices.csv')
df = pd.read_csv(data_path)
display(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
new_df = df.copy()
new_df = new_df.drop('customerID', axis=1)
```

```python
cat_col = new_df.select_dtypes(exclude='number').columns
for cat in cat_col:
    plt.figure(figsize=(12,5))
    plt.title(f'distribuiÃ§ao da coluna {cat}')
    sns.countplot(data=df, x=df[cat], palette='viridis', saturation=.5)
    plt.xlabel(f'{cat}')
    plt.ylabel('contagem')
    plt.tight_layout()
    plt.show()
```

```python
for cat_churn in cat_col:
    plt.figure(figsize=(12,5))
    plt.title(f'distribuiÃ§ao da coluna {cat_churn}')
    sns.countplot(data=df, x=df[cat_churn], palette='viridis', saturation=.5, hue='Churn')
    plt.xlabel(f'{cat_churn}')
    plt.ylabel('contagem')
    plt.tight_layout()
    plt.show()
```
